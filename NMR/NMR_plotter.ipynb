```python
#Initiate libraries needed for script
import glob, os
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib as matplotlib
import seaborn as sns
import matplotlib.cm as cm
from numpy import linspace
import statistics
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import scipy
from scipy.signal import find_peaks
from scipy.stats import linregress

#functions used in this script:

#opening directory and grabbing all text files with wildcard
#in this case it is all with DMPC_LUV in the name
#function to pull all files (X and Y coordinates) for text files with wildcard matching input
def csv_pull(files, cols, trim):
    li = []
    f = sorted(glob.glob(files))
    file_titles = []
    for n in f:
        df = pd.read_csv(n, skiprows = 1, sep=",", names = cols)
        temp = n[trim[0]:trim[1]]
        #temp = n.strip(files)
        file_titles.append(temp)
        li.append(df)
        
    return li, file_titles

#Finds the location of the peaks between expected ranges for -PC and -PG headgroups
#As the series is done in triplicate, finds this for each of the readings, then averages them with standard deviation
def peaking(arr,range_):
    peak_collection = []
    for n in arr:
        #find the range to look for peaks
        d = []
        temp = np.where(round(n['ppm'],1)== range_[1])
        d.append(temp[0][0])
        temp = np.where(round(n['ppm'],1)== range_[0])
        d.append(temp[0][0])
        #find the peaks between these ranges
        peaks, _ = find_peaks(n['int'][d[0]:d[1]],height= 0.9*np.max(n['int'][d[0]:d[1]]))
        peaks = peaks+d[0]
        peak_collection.append(n['ppm'][peaks[0]])
    #average all peak ppm collected and give errors
    avg = round(np.mean(peak_collection),5)
    err = round(np.std(peak_collection),5)
    
    #Collate and return the findings
    peak_info = [peak_collection, avg, err]
    return peak_info


#Finding %lipid mixed:
#b = the peak locations for the sample (in triplicate so A,B,C)
#standard = the % of lipid used for the standards (100,75,50,25)
#values = The lipid standard peak locations (i.e. the peaks for 100, 75, 50, 25)
def mixer_finder(b, standard, values):
    #find the slope, intercept etc fr the standard titrations
    a = []
    slope, intercept, r_value, p_value, std_err = linregress(standard,values)
    
    #If the peak locations is an array, this will run it in a loop and return an array back
    if b == float:
        #Looping!
        for n in b:
            #Calculate the x (% of lipid mixing) from the y (the ppm of the peak)
            n = n-intercept
            n = n/slope
            a.append(n)
    #If it is a single spectra, it will return the value for that single spectra
    else:
        b = b-intercept
        b = b/slope
        a = b
    return a

#Standard colours used within the thesis
colors = [sns.xkcd_rgb["black"], #0, black
          sns.xkcd_rgb["cerulean"], #1 blue
          sns.xkcd_rgb["moss green"], #2 dark green
          sns.xkcd_rgb["rust"], #3 dark red
          sns.xkcd_rgb["slate"], #4 slate/grey-blue
          sns.xkcd_rgb["mustard"], #5 yellow
          sns.xkcd_rgb["tangerine"], # 6 tangerine ***** this is for SMA 3:1
          sns.xkcd_rgb["coral"], #7 pink-red ***** This is for SMA 2:1
          sns.xkcd_rgb["grass"], #8 grass (brighter green)***** This is for DIBMA
          sns.xkcd_rgb["bright blue"], #9  bright blue *****This is for PMA
          sns.xkcd_rgb["olive green"], #10 olive green
          sns.xkcd_rgb["pastel purple"] #11 pastel purple
         ]
#standard plotting style used within this thesis.
sns.set_style("white")
sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
#Read in the files, with standard column headers and trimmings
cols = ['pts','int','x','ppm']
trim = [0,8]
PC_100, files = csv_pull('100PC*.txt', cols, trim)
PC_75, files = csv_pull('75PC*.txt', cols, trim)
PC_50, files = csv_pull('50PC*.txt', cols, trim)
PC_25, files = csv_pull('25PC*.txt', cols, trim)
PG_100, files = csv_pull('100PG*.txt', cols, trim)
sample, files = csv_pull('Sample*.txt', cols, trim)

#this is the manually chosen range, as you can see they're quite broad.
#The only need forthe range is actually to ensure they are both detected, 
#if not I found that sometimes it swaps the peaks round
PC_range = [-1, .3]
PG_range = [.8,1.5]
#Empty rrays ready to receive the data need to initialised here
PC_peaks = []
PC_err = []
PG_peaks = []
PG_err = []
#Calls the peaking function to find the peak data from each spectra (averaged)
#For each sample there aretwo reports, one for each lipid head group
#For clarity, these are labelled in % of lipid in the sample, so PC_75 is the PC peak of the 75% DMPC:25% DMPG sample
#spits out the peaks ([0]), peak average [1], peak error [2]

#DMPC peaks for all controls
peaks_ = peaking(PC_100, PC_range)
PC_peaks.append(peaks_[1])
PC_err.append(peaks_[2])

peaks_ = peaking(PC_75, PC_range)
PC_peaks.append(peaks_[1])
PC_err.append(peaks_[2])

peaks_ = peaking(PC_50, PC_range)
PC_peaks.append(peaks_[1])
PC_err.append(peaks_[2])

peaks_ = peaking(PC_25, PC_range)
PC_peaks.append(peaks_[1])
PC_err.append(peaks_[2])

#now onto the -PG peaks
peaks_ = peaking(PG_100, PG_range)
PG_peaks.append(peaks_[1])
PG_err.append(peaks_[2])

peaks_ = peaking(PC_25, PG_range)
PG_peaks.append(peaks_[1])
PG_err.append(peaks_[2])

peaks_ = peaking(PC_50, PG_range)
PG_peaks.append(peaks_[1])
PG_err.append(peaks_[2])

peaks_ = peaking(PC_75, PG_range)
PG_peaks.append(peaks_[1])
PG_err.append(peaks_[2])


#This is for the actual measured sample, the above were the titration controls
PC_sample_peaks = peaking(sample, PC_range)
PG_sample_peaks = peaking(sample, PG_range)
```

```python
#Plotting Figure, and generating an individual colour from the Spectral colour gradient
fig, axs = plt.subplots(6, 1, figsize=(18, 18), frameon = True, sharex = True)
coloring = [cm.Spectral(x) for x in linspace(0, 1, 6)]
i = 0

#plot graphs
#as there are multiple spectra per control/sample, it is easier to do this in a loop all at once than individually
while i < len(files):
    axs[0].plot(sample[i]['ppm'], sample[i]['int'], color = colors[0])
    axs[1].plot(PC_100[i]['ppm'], PC_100[i]['int'], color = coloring[0])
    axs[2].plot(PC_75[i]['ppm'], PC_75[i]['int'], color = coloring[1])
    axs[3].plot(PC_50[i]['ppm'], PC_50[i]['int'], color = coloring[3])
    axs[4].plot(PC_25[i]['ppm'], PC_25[i]['int'], color = coloring[4])
    axs[5].plot(PG_100[i]['ppm'], PG_100[i]['int'], color = coloring[5])

    i = i+1
i =0
#things for all axis (title, axis widths etc)
title = ["Sample",'100% PC', 'PC:PG 75:25', 'PC:PG 50:50', 'PC:PG 25:75', '100% PG']
labs = ["A","B","C","D","E","F"]
#Standardise the axes for visual comparison
#min and max here are x axis limits
min_ = -.2
max_ = 1.42
#Setting the ctual limits in a loop as 6 different plots
for n in axs:
    n.set_xlim(max_,min_)
    #The y_lim was done manually, I'm sure there is an easier, neater, way than this but oh well!
    n.set_ylim(-20000000, 900000000)
    n.set_yticks([])
    n.text(x = max_, y = 750000000, s = title[i], fontsize = 13)
    
    #set labels
    height = n.axis()
    position = height[0]-(0.02*(height[1]- height[0]))
    n.text(x = position, y = 0.9*height[3], s = labs[i], weight = "bold", fontsize = 20)
    i = i+1
    

#annotate the peaks (PC only)
axs[0].text(x = PC_sample_peaks[1]-0.04, y = 0.9*max(sample[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PC_sample_peaks[1],PC_sample_peaks[2]), fontsize = 13, color = colors[0])
axs[1].text(x = PC_peaks[0]-0.04, y = 0.9*max(PC_100[0]['int']), 
            s = "{:.3f} "u"\u00B1{:.3f}".format(PC_peaks[0], PC_err[0]), fontsize = 13, color = colors[0])
axs[2].text(x = PC_peaks[1]-0.04, y = 0.9*max(PC_75[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PC_peaks[1], PC_err[1]), fontsize = 13, color = colors[0])
axs[3].text(x = PC_peaks[2]-0.04, y = 0.9*max(PC_50[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PC_peaks[2], PC_err[2]), fontsize = 13, color = colors[0])
axs[4].text(x = PC_peaks[3]-0.04, y = 0.9*max(PC_25[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PC_peaks[3], PC_err[3]), fontsize = 13, color = colors[0])


#annotate peaks (PG only)
axs[0].text(x = PG_sample_peaks[1]-0.04, y = 0.9*max(sample[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PG_sample_peaks[1],PG_sample_peaks[2]), fontsize = 13, color = colors[0])
axs[2].text(x = PG_peaks[3]-0.04, y = 0.9*max(PC_75[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PG_peaks[3], PG_err[3]), fontsize = 13, color = colors[0])
axs[3].text(x = PG_peaks[2]-0.04, y = 0.9*max(PC_50[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PG_peaks[2],PC_err[2]), fontsize = 13, color = colors[0])
axs[4].text(x = PG_peaks[1]-0.04, y = 0.9*max(PC_25[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PG_peaks[1],PG_err[1]), fontsize = 13, color = colors[0])
axs[5].text(x = PG_peaks[0]-0.04, y = 0.9*max(PG_100[0]['int']),
            s = "{:.3f} "u"\u00B1{:.3f}".format(PG_peaks[0],PG_err[0]), fontsize = 13, color = colors[0])
#set the shared x-axis label
axs[5].set_xlabel('ppm')
#Save and plot!
sns.despine(left = True, top  = True, bottom = True)
plt.savefig('NMR_spectra.png', format='png', dpi=300, bbox_inches="tight")
plt.show()
```

```python
#The plotting of the standard curve

#These are the x-axis (i.e. the % of that lipid) requird to plot the titration points
PC = [100, 75, 50, 25]
PG = [100, 75, 50, 25]

#Two plots, one for DMPC and one for DMPG
fig, axs = plt.subplots(2, 1, figsize=(15, 8), frameon = True, sharex = True)
#DMPC peaks
#per plot there are two components, the error bar plot for the actual data (averaged and error-ed)
axs[0].errorbar(PC, PC_peaks, yerr = PC_err, color = colors[0], capthick = 1, capsize = 5,
             lw = 2,  marker = "o", markeredgecolor = colors[0], markerfacecolor = colors[0],
             markeredgewidth = 1.5, markersize = 8, label = "DMPC")
#This plot however is the linear regression (i.e. the straight line fit)
axs[0].plot(np.unique(PC), np.poly1d(np.polyfit(PC, PC_peaks, 1))(np.unique(PC)), color = colors[0], ls = ":")

#DMPG peaks
axs[1].errorbar(PG, PG_peaks, yerr = PG_err, color = colors[1], capthick = 1, capsize = 5,
             lw = 2,  marker = "o", markeredgecolor = colors[0], markerfacecolor = colors[1],
             markeredgewidth = 1.5, markersize = 8, label = "DMPG")
axs[1].plot(np.unique(PG), np.poly1d(np.polyfit(PG, PG_peaks, 1))(np.unique(PG)), color = colors[1], ls = ":")

#Labelling the axes
axs[1].set_xlabel("% of Lipid")
axs[0].text(x = 14, y = 0, s="Chemical Shift (ppm)", rotation= 90)

height = axs[0].axis()
position = height[0]-(0.04*(height[1]- height[0]))
axs[0].text(x = position, y = 1*height[3], s = "A", weight = "bold", fontsize = 20)
height = axs[1].axis()
position = height[0]-(0.04*(height[1]- height[0]))
axs[1].text(x = position, y = 1*height[3], s = "B", weight = "bold", fontsize = 20)

axs[0].set_title("DMPC", fontsize = 13, loc = 'left')
axs[1].set_title("DMPG", fontsize = 13, loc = 'left')

sns.despine()
plt.show()
```

```python
#The above is a visual check of the linear regression (straight line fit) for the titration points
#from this, we can now deduce the % of mixing in each of the sample triplicates

#Using the function mixer_finder
#To obtain the errors, mixer_finder works for each triplicate sample, then returns them as a 3x1 array
PC_sample = mixer_finder(PC_sample_peaks[0], PC, PC_peaks)
#The 3x1 array  is then averaged and standard errored
PC_x_err = np.std(PC_sample)
PC_x_avg = np.mean(PC_sample)
print(PC_x_avg)
print(PC_x_err)

#this is done for each lipid type
PG_sample = mixer_finder(PG_sample_peaks[0], PG, PG_peaks)
PG_x_err = np.std(PG_sample)
PG_x_avg = np.mean(PG_sample)
print(PG_x_avg)
print(PG_x_err)
```
