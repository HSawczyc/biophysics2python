```python
#Initiate libraries needed for script
import glob
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib
import seaborn as sns
import matplotlib.cm as cm
from numpy import linspace
import copy
import statistics
import matplotlib.gridspec as gridspec

#functions used in this script:

#opening directory and grabbing all text files with wildcard
#function to pull all files (X and Y coordinates) for text files with wildcard matching input
def filepull(a,p):
    files = []
    for file in sorted(glob.glob(a)):
        files.append(file)

    fx = []
    fy = []

    #For each text file in the directory, opening it and grabbing the plotting data within the file
    for fn in files:
        temp_a, temp_x, temp_y = np.loadtxt(fn, unpack = True, skiprows = 2)
        fx.append(temp_x)
        fy.append(temp_y)
    #print(files)
    #putting into unique array for calling later on to plot
    x = np.asarray(fx)
    y = np.asarray(fy)

    if p == 1:
        print(files)
    return x,y

#Calculates the order parameter (S) used in the thesis
def S_calc(A_pa, A_pe):
    #A factors are in Gauss, from Knight et al. 1981 pg 147
    Axx = 5.9
    Ayy = 5.4
    Azz = 32.9
    
    a_prime_0 = (1/3)*(Azz+Ayy+Axx)
    a_0 = (1/3)*(A_pa+(2*A_pe))
    a_temp = a_prime_0/a_0
    
    temp1 = A_pa-A_pe
    temp2 = (Azz-(1/2)*(Axx+Ayy))
    S_temp = temp1/temp2
    S = S_temp*a_temp
    return S

#This finds the A// positions from the high point (left side of the spectra)
#and the lowest point (right side of the spectra)
#The ranges given were defined manually (roughly) to avoid mis labelling
def Aii_find(x,y,rangestart,rangeend):
    length = len(x)
    i = 0
    a_peak = min(y)
    a_trough = max(y)
    while i < length:
        if x[i] > rangestart[0] and x[i] < rangestart[1]:
            if a_peak < y[i]:
                a_peak = y[i]
                a_start = x[i]
        
        if x[i] > rangeend[0] and x[i] < rangeend[1]:
            if a_trough > y[i]:
                a_trough = y[i]
                a_end = x[i]
        i = i+1
    #print(a_start, a_end)
    a = a_end - a_start
    
    a = (a/2)
    return a

#This finds the A perpendicular positions from the low point (left side of the spectra)
# and the high point (right side of the spectra)
#ranges are, again, chosen manually to avoid mis labelling
def ap_find(x,y,rangestart,rangeend):
    length = len(x)
    i = 0
    a_peak = max(y)
    a_trough = min(y)
    
    while i < length:
        if x[i] > rangestart[0] and x[i] < rangestart[1]:
            #print(a_peak,y[i])
            if a_peak == 0:
                a_peak = y[i]
                
            if a_peak >= y[i]:
                a_peak = y[i]
                a_start = x[i]
        
        if x[i] > rangeend[0] and x[i] < rangeend[1]:
            if a_trough <= y[i]:
                a_trough = y[i]
                a_end = x[i]
        i = i+1
    #print(a_start, a_end)
    a = a_end - a_start
    a = (a/2)
    return a

#Quick script to calculate the order paramter for all temperatures tested for all samples with the three above functions
def order_calc_many(x, y, start, fake_peak):
    A_ll = [[3315,3325],[3365,3385]]
    A_pe = [[3330,3340],[3347,3360]] #normally 3360
    S = []
    i = 0
    while i < len(x):
        #Due to difficulty in automatic detection of A_pe 2nd peak, a manual 'fake' peak was chosen for these spectra
        if i < start:
            S.append(S_calc(Aii_find(x[i], y[i], A_ll[0], A_ll[1]), ap_find(x[i], y[i], A_pe[0], fake_peak)))
            #after this is reached, automatic detection can resume
        else:
            S.append(S_calc(Aii_find(x[i], y[i], A_ll[0], A_ll[1]), ap_find(x[i], y[i], A_pe[0], A_pe[1])))
        i= i+1
        
    return S
    
#Standard colours used within the thesis
colors = [sns.xkcd_rgb["black"], #0, black
          sns.xkcd_rgb["cerulean"], #1 blue
          sns.xkcd_rgb["moss green"], #2 dark green
          sns.xkcd_rgb["rust"], #3 dark red
          sns.xkcd_rgb["slate"], #4 slate/grey-blue
          sns.xkcd_rgb["mustard"], #5 yellow
          sns.xkcd_rgb["tangerine"], # 6 tangerine ***** this is for SMA 3:1
          sns.xkcd_rgb["coral"], #7 pink-red ***** This is for SMA 2:1
          sns.xkcd_rgb["grass"], #8 grass (brighter green)***** This is for DIBMA
          sns.xkcd_rgb["bright blue"], #9  bright blue *****This is for PMA
          sns.xkcd_rgb["olive green"] #10 olive green
         ]

#standard plotting style used within this thesis.
sns.set_style("white")
sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
#due to the frozen spectra possessing a very weak A_pe second peak, 
#a manually selected counter for each sample was selected
#before this is reached, the spectra peak can not be found by the script so a narrower range is given
#After this, the script measures the highest point as it can find 


#spectra from which the script can automatically detect the A_pe scond peak
LUV_start = 14
#This range is the A_pe 2nd peak range to detect
#these were manually selected for each sample type due to the variation, and sloped nature of the spectra in this region
LUV_peak = [3344,3355]
LUV_S = order_calc_many(LUVx, LUVy, LUV_start, LUV_peak)
SMA31_start = 13
SMA31_peak = [3347,3352]
SMA31_S = order_calc_many(SMA31x, SMA31y, SMA31_start, SMA31_peak)
SMA21_start = 13 
SMA21_peak = [3347,3352]
SMA21_S = order_calc_many(SMA21x, SMA21y, SMA21_start, SMA21_peak)
DIBMA_start = 11
DIBMA_peak = [3348,3352.9]
DIBMA_S = order_calc_many(DIBMAx, DIBMAy, DIBMA_start, DIBMA_peak)
PMA_start = 10
PMA_peak = [3347,3352]
PMA_S = order_calc_many(PMAx, PMAy, PMA_start, PMA_peak)

#Ok, time to plot!
```
